```python
 from Bio import SeqIO

 proteome_file = "/content/uniprot_proteome.fasta"
with open(proteome_file, "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        description = record.description
        # Extract gene name and ID from the description
        print(description)
```

```python
gene_list = ["GENE1","ABAT","ABCA1","ABCB1","ABCB11","ABCC1","ABCC8","ABCC9","ABCG2","ABL1","ABL2","ACAA2","ACACB","ACADSB","ACAT1"]  # Replace with your target genes
for record in SeqIO.parse(proteome_file,"fasta"):
    description = record.description
    for gene in gene_list:
        if gene in description:
            print(f"Found: {description}")
```
